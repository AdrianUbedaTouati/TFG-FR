```python
# Mostrar tipo de objeto principal
print("Tipo del objeto 'df':", type(df))

# Ver las dimensiones del DataFrame
print(f"\nEl DataFrame tiene {df.shape[0]} filas y {df.shape[1]} columnas.")

# Ver nombres de columnas y sus tipos
print("\nTipos de datos por columna:\n")
print(df.dtypes)

# Ver conteo de valores no nulos por columna
print("\nConteo de valores no nulos por columna:\n")
print(df.count())

# Describir brevemente cada tipo de columna
print("\nResumen por tipo:")
print(f"- Columnas numéricas: {df.select_dtypes(include='number').shape[1]}")
print(f"- Columnas categóricas: {df.select_dtypes(include='object').shape[1]}")
print(f"- Columnas con valores faltantes: {(df.isnull().sum() > 0).sum()}")
```
